### Logistic Regression for Handwritting Recognition

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x,y = load_digits(return_X_y = True)
```

```python
print(x,'\n',y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25,random_state=2024)
```

```python
x_train
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = LogisticRegression(solver='liblinear',C=0.05,multi_class='ovr',random_state=2024)
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_train,y_pred_train)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
cm = confusion_matrix(y_test,y_pred_test)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
print('Train Set: \n',classification_report(y_train,y_pred_train))

print('Test Set: \n',classification_report(y_test,y_pred_test))
```

## Parametric vs Non-Parametric Models

### KNN

```python
#Feature 1
weather = ['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny','Rainy',
          'Sunny','Overcast','Overcast','Rainy']
# Feature 2
temp = ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool',
       'Mild','Mild','Mild','Hot','Mild']

# Label atau target
play = ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']

print(len(weather),len(temp),len(play))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
weather_enc = le.fit_transform(weather)
temp_enc = le.fit_transform(temp)
play_enc = le.fit_transform(play)
print(weather_enc)
print(temp_enc)
print(play_enc)
```

```python
features = list(zip(weather_enc,temp_enc))
features
```

```python
from sklearn.neighbors import KNeighborsClassifier

x = features
y = play_enc
model = KNeighborsClassifier(n_neighbors=3)
model.fit(x,y)
```

```python
predicted = model.predict([[0,1]])
print(predicted)
```

```python
y_pred = model.predict(x)
```

```python
print(classification_report(y,y_pred))
```

### KNN with Multiple Labels

```python
from sklearn.datasets import load_wine
```

```python
wine['feature_names']
```

```python
wine['data']
```

```python
wine['target']
```

```python
wine['data'].shape
```

```python
import pandas as pd
```

```python
x = pd.DataFrame(wine['data'],columns=wine['feature_names'])
x
```

```python
y= pd.DataFrame(wine['target'],columns=['kultivar'])
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=2024)
```

```python
knn = KNeighborsClassifier(n_neighbors=5) # K = 5
knn.fit(X_train,y_train)

y_pred = knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

print('accuracy:', accuracy_score(y_test,y_pred))
```

```python
knn_7 = KNeighborsClassifier(n_neighbors=7)
knn_7.fit(X_train,y_train)

y_pred_7 = knn_7.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('accuracy:', accuracy_score(y_test,y_pred_7))
```

```python
y_pred != y_test['kultivar']
```

```python
#kalkulasi error untuk k-values 1-40

error = []

for i in range (1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error.append(np.mean(pred_i != y_test['kultivar']))
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(1,40),error,color='red',marker='o')
plt.title('Error rate K')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn_11 = KNeighborsClassifier(n_neighbors=11)
knn_11.fit(X_train,y_train)

y_pred_11 = knn_11.predict(X_test)

print('Accuracy:', accuracy_score(y_test,y_pred_11))
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn_11 = KNeighborsClassifier(n_neighbors=11)
knn_11.fit(X_train_scaled,y_train)
y_pred_11_scaled = knn_11.predict(X_test_scaled)
print('Accuracy:',accuracy_score(y_test,y_pred_11_scaled))
```

```python
y_pred_train = knn_11.predict(X_train_scaled)
y_pred_test = knn_11.predict(X_test_scaled)
```

```python
print('Train Set')
print(classification_report(y_train,y_pred_train))
print('Test Set')
print(classification_report(y_test,y_pred_test))
```
