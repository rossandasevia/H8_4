## Pie Chart

Pie chart menggambarkan proporsi dari data

```python
df_can.head()
```

```python
#preparasi data piechart
df_continents = df_can.groupby('Continent').sum()
df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize=(5,6),
                           autopct='%1.1f%%',
                           startangle=90,
                           shadow=True)
plt.title('Immigration to Canada by Continent')
plt.axis('equal')
plt.show
```

```python
#mengubah warna chart

colors_list = ['gold','yellow','lightgreen','lightcoral','pink','lightskyblue']
explode_list = [0.1,0,0,0,0.05,0.15]

df_continents['2013'].plot(kind='pie',
                          figsize=(15,6),
                          autopct='%1.1f%%',
                          startangle=90,
                          shadow=True,
                          labels=None,
                          pctdistance=1.12,
                          colors=colors_list,
                          explode=explode_list)
plt.legend(labels=df_continents.index,loc='upper left')
plt.axis('equal')
plt.title('Immigration to Canada by Continent 2013',y=1.12) # y digunakan untuk meninggikan judul jika overlap dengan teks pada chart
plt.show()
```

## WordClouds

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
alice_novel = open('alice_novel.txt','r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
alice_wc = WordCloud(background_color='white',
                    max_words=2000,
                    stopwords=stopwords)

#generate word cloud
alice_wc.generate(alice_novel)
```

```python
fig = plt.figure(figsize=(8,5))
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```

```python
df_incidents=pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Police_Department_Incidents_-_Previous_Year__2016_.csv')
df_incidents.head()
```

```python
#mapping 100 rows pertama dari tabel
df_incidents_100 = df_incidents.iloc[0:100,:]
df_incidents_100
```

```python
#San Fransisco Lat and Lon
lat= 37.77
lon=-122.42
```

```python
sf_map = folium.Map(location=[lat,lon])
sf_map
```

```python
incidents = folium.map.FeatureGroup()

#add 100 crimes to FeatureGroup

for lat,lon in zip(df_incidents_100.Y,df_incidents_100.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat,lon],
            radius=5,
            color='blue',
            fill=True,
            fill_color='red',
            fill_opacity=0.6
        )
    )
sf_map.add_child(incidents)
```
